```python
# import pipeline
from sklearn.pipeline import Pipeline
#import standaer scalar
from sklearn.preprocessing import StandardScaler
# import simple imputer
from sklearn.impute import SimpleImputer 
```

```python
#skewness handling
# create pipeline
# 1.simple imputer
#missing_values type for imputation
imputer = SimpleImputer(missing_values = np.nan,
                        strategy ='mean')

# Fitting the data to the imputer object
imputer = imputer.fit(X)
 
# Imputing the data   
X = imputer.transform(X)
 
print("Imputed Data : \n", X)
# 2. standard scalar 
object= StandardScaler() 
# standardization 
scale = object.fit_transform(X) 
print(scale)
```

```python
# import train test split
from sklearn.model_selection import train_test_split
```

```python
# Split into training (80%) and testing set (20%)
train_x, test_x, train_y, test_y= train_test_split(X,y,test_size=0.3,random_state=1)

# check the shape of train and test
train_x.shape,train_y.shape,test_x.shape, test_y.shape
```

```python
# import SMOTE module from imblearn library
#SMOTE library is used to give the extra weight to minority data to balace the target variable
from imblearn.over_sampling import SMOTE
# Instantiate
#from imblearn.datasets import make_imbalance
#fit on train data

smote = SMOTE(random_state = 2)
X_train_res, y_train_res = smote.fit_resample(train_x, train_y.ravel())

print('After OverSampling, the shape of train_X: {}'.format(X_train_res.shape))
print('After OverSampling, the shape of train_y: {} \n'.format(y_train_res.shape))
  
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res == 1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res == 0)))
# printthe shape of X_train After OverSampling

# printthe shape of y_train After OverSampling
```
