```python
# import pipeline
from sklearn.pipeline import Pipeline
#import standaer scalar
from sklearn.preprocessing import StandardScaler
# import simple imputer
from sklearn.impute import SimpleImputer 
```

```python

#missing_values type for imputation
imputer = SimpleImputer(missing_values = np.nan,
                        strategy ='mean')

# Fitting the data to the imputer object
imputer = imputer.fit(X)
 
# Imputing the data   
X = imputer.transform(X)
 
# print("Imputed Data : \n", X)
# # 2. standard scalar 
# object= StandardScaler() 
# # standardization 
# scale = object.fit_transform(X) 
# print(scale)
```

```python
# import train test split
from sklearn.model_selection import train_test_split
```

```python
# Split into training (80%) and testing set (20%)
train_x, test_x, train_y, test_y= train_test_split(X,y,test_size=0.2,random_state=41)

# check the shape of train and test
train_x.shape,train_y.shape,test_x.shape, test_y.shape
```

```python
#from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
#ros = RandomOverSampler(sampling_strategy='not minority') # Float
ros = RandomUnderSampler(sampling_strategy=1) # String
X_train_res, y_train_res = ros.fit_resample(train_x, train_y)
```

```python
# importing necessary libraries for geting metrics of models
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score,roc_auc_score,roc_curve
```

```python

# rf=RandomForestClassifier(n_estimators=300, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1,min_weight_fraction_leaf=0.0,
#                                       max_features='sqrt', max_leaf_nodes=None,
#                                       bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False,
#                                       class_weight=None, ccp_alpha=0.0, max_samples=None)
```

```python
# ##implement the XGBoost classifier

from xgboost import XGBClassifier


# declare parameters
params = {
            
            'max_depth': 5,
            'alpha': 10,
            'learning_rate': 0.1,
            'n_estimators':250
        }
            
            
            
# instantiate the classifier 
xgb_clf = XGBClassifier(**params)



# fit the classifier to the training data
xgb_clf.fit(X_train_res, y_train_res)


```

```python
#rf.fit(X_train_res,y_train_res)

# Predicting the test set results
y_pred = xgb_clf.predict(test_x)
#y_pred = rf.predict(test_x)
ACC=(accuracy_score(test_y, y_pred))
print(ACC)
print(classification_report(test_y,y_pred))
```

```python
#plotting the ROC and AUC
roc_auc = roc_auc_score(test_y, y_pred)

false_positive_rate, true_positive_rate, thresolds =roc_curve(test_y, y_pred)

plt.figure(figsize=(10, 8), dpi=80)
plt.axis('scaled')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title("AUC & ROC Curve")
plt.plot(false_positive_rate, true_positive_rate, 'g')
plt.fill_between(false_positive_rate, true_positive_rate, facecolor='lightgreen', alpha=0.7)
plt.text(0.95, 0.05, 'AUC = %0.4f' % roc_auc, ha='right', fontsize=10, weight='bold', color='blue')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.show()
```
